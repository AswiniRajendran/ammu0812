# 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
from matplotlib import pyplot as plt
buyers = cars_data[['Buyer Gender']]
buyers.describe() # use describe function to find out the male- female count of buyers
```

```python
plt.bar(['Female','Male'],[5052,4948])
plt.xlabel("Gender")
plt.ylabel("Frequency of Buyers")
plt.title("Female vs. Male Buyers on Sales")
plt.show()
# There is only a slight difference in these two values
```

# 4. Find the top 5 cars based on their sales price.

```python
# Sort the data in descending order of Sale Price and display the data. This gives us the top 5 cars based on Sales Price
cars_data.sort_values("Sale Price", axis = 0, ascending = False, inplace = True, na_position = 'last')
```

```python
cars_data
```

```python
# Hence the top 5 cars are, 
# 1. Volkswagen-Jetta III
# 2. Audi-A6
# 3. Lexus-LX
# 4. Mazda-MX6
# 5. Lexus-ES
```

# 5. Find the least 5 cars based on their Resell price.

```python
# Sort the data in ascending order of Resell Price and display the data. This gives us the least 5 cars based on Resell Price.
cars_data.sort_values("Resell Price", axis = 0, ascending = True, inplace = True, na_position = 'last')
```
